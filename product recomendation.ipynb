```python
snapshot_date = df['InvoiceDate'].max() + pd.Timedelta(days=1)
rfm = df.groupby('CustomerID').agg({
    'InvoiceDate': lambda x: (snapshot_date - x.max()).days,
    'InvoiceNo': 'nunique',
    'TotalPrice': lambda x: (df.loc[x.index, 'Quantity'] * df.loc[x.index, 'UnitPrice']).sum()
})
rfm.columns = ['Recency', 'Frequency', 'Monetary']
```

```python
rfm.head()
```

```python
plt.figure(figsize=(15,4))
for i, column in enumerate(['Recency', 'Frequency', 'Monetary']):
    plt.subplot(1, 3, i+1)
    sns.histplot(rfm[column], kde=True, bins=30)
    plt.title(column)
plt.tight_layout()
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
rfm_scaled = scaler.fit_transform(rfm[['Recency', 'Frequency', 'Monetary']])
```

```python
from sklearn.cluster import KMeans
# Elbow method
inertia = []
K_range = range(1, 11)

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(rfm_scaled)
    inertia.append(kmeans.inertia_)

# Plot Elbow
plt.figure(figsize=(8, 5))
plt.plot(K_range, inertia, marker='o')
plt.title('Elbow Method for Optimal K')
plt.xlabel('Number of clusters (K)')
plt.ylabel('Inertia (Within-cluster sum of squares)')
plt.grid(True)
plt.show()
```

```python
kmeans = KMeans(n_clusters=4, random_state=42)
rfm['Cluster'] = kmeans.fit_predict(rfm_scaled)
```

```python
rfm = rfm.reset_index() 
```

```python
cluster_summary = rfm.groupby('Cluster').agg({
    'Recency': 'mean',
    'Frequency': 'mean',
    'Monetary': 'mean',
    'CustomerID': 'count'
}).rename(columns={'CustomerID': 'NumCustomers'}).round(2)

cluster_summary
```

```python
def label_cluster(row):
    if row['Recency'] < 20 and row['Frequency'] > 10 and row['Monetary'] > 1000:
        return 'High-Value'
    elif row['Recency'] < 50 and row['Frequency'] > 5:
        return 'Regular'
    elif row['Recency'] > 60 and row['Frequency'] < 3:
        return 'At-Risk'
    else:
        return 'Occasional'

rfm['Segment'] = rfm.apply(label_cluster, axis=1)
```

####  Product Recommendation System

```python
df_reco = df[['CustomerID', 'StockCode', 'Description', 'Quantity']].copy()
```

```python
customer_product_matrix = df_reco.pivot_table(index='CustomerID',
                                              columns='StockCode',
                                              values='Quantity',
                                              aggfunc='sum').fillna(0)
```

```python
customer_product_matrix.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
product_similarity = cosine_similarity(customer_product_matrix.T)
```

```python
similarity_df = pd.DataFrame(product_similarity,
                             index=customer_product_matrix.columns,
                             columns=customer_product_matrix.columns)
```

```python
# Build a product name -> StockCode lookup dictionary
product_lookup = df[['StockCode', 'Description']].drop_duplicates().dropna()

# Drop generic or null descriptions
product_lookup = product_lookup[~product_lookup['Description'].str.contains('?', regex=False)]

# Create reverse lookup dictionary
desc_to_code = product_lookup.drop_duplicates('Description').set_index('Description')['StockCode'].to_dict()
code_to_desc = product_lookup.drop_duplicates('StockCode').set_index('StockCode')['Description'].to_dict()
```

```python
def get_similar_products(product_name, top_n=5):
    try:
        # Get StockCode from product name
        product_code = desc_to_code[product_name]

        # Get similar product codes
        sim_scores = similarity_df[product_code].sort_values(ascending=False)[1:top_n+1]

        # Map codes back to descriptions
        similar_descriptions = [code_to_desc[code] for code in sim_scores.index if code in code_to_desc]

        return similar_descriptions
    
    except KeyError:
        return ["Product not found. Please check the name."]
```
